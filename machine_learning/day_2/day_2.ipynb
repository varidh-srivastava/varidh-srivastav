```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv('../input/titanic/tested.csv')
data.head()
#data from titanic dataset
```

```python
sns.heatmap(data.isnull(),yticklabels=False,cbar=False,cmap='viridis')
#A visulaization showing description of age and cabin(class of board.)
```

```python

sns.countplot(x='Survived',data=data)
#This is plot showing number of people survived in incident or died(0->died,1->survived)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='Survived',hue='Sex',data=data)
```

```python
sns.countplot(x='Survived',hue='Pclass',data=data)

#this visualization shows that number of people survived according to the board-class they booked
```

```python
sns.distplot(data['Age'].dropna(),bins=40)
#number of people present in the ship according to their age;
```

```python
sns.boxplot(x='Survived',y='Age',data=data)
#plot to show survived according to their age;
```
